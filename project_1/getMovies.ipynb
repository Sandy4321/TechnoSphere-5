```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'http://www.imdb.com/chart/bottom'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
titleColumn = soup.find_all('td', class_='titleColumn')
posterColumn = soup.find_all('td', class_='posterColumn')
ratingColumn = soup.find_all('td', class_='ratingColumn imdbRating')
```

```python
movieList = pd.DataFrame(columns=['name', 'id', 'year', 'director', 'top', 'rating', 'num_of_rates'])
```

```python
for i in xrange(250):
    movie = {}
    movie['top'] = i+1
    movie['name'] = titleColumn[i].find('a').get_text()
    #movie['director'] = titleColumn[i].find('a').atrtrs['title']
    movie['year'] = int(titleColumn[i].find('span').get_text()[1:5])
    movie['rating'] = float(ratingColumn[i].find('strong').attrs['title'].split(' ')[0])
    movie['num_of_rates'] = int(ratingColumn[i].find('strong').attrs['title'].split(' ')[3].replace(",",""))
    movie['id'] = posterColumn[i].find('a').attrs['href'].split('/')[2]
    movieList = movieList.append(movie,ignore_index=True)
movieList = movieList.drop_duplicates()
```

```python
movieList.to_csv('./movieList.csv', encoding='utf-8', index=False)
```
